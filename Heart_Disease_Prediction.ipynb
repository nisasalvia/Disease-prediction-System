```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('/content/dataset_heart.csv')
```

```python
X = data.drop(columns='heart disease', axis=1)
Y = data['heart disease']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, stratify = Y, random_state=2)
```

```python
print(X.shape , X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
pred = model.predict(X_train)
data_acc = accuracy_score(pred, Y_train)
```

```python
print(pred)
```

```python
print(data_acc)
```

```python
pred_test = model.predict(X_test)
data_testacc = accuracy_score(pred_test, Y_test)
```

```python
print(data_testacc)
```

```python
input_data = (70,1,4,130,322,0,2,109,0,2.4,2,3,3)
data_as_numpy = np.asarray(input_data)
data_reshaped = data_as_numpy.reshape(1,-1)
prediction = model.predict(data_reshaped)
print(prediction)
if(prediction[0] == 1):
  print("The person do not have Heart Disease.")
else:
  print("The person has heart disease.")
```
